```python
def read_file(path):
    with open(path, "r") as readfile:
        a = readfile.read()
        lines = list(a.split("\n"))
    return list(lines)
```

```python
def input_lists_modifier(input):
    new_input = []
    for line in input:
        new_input.append(list(line))
    
    return new_input
```

```python
def start_finder(input):
    for ix, line in enumerate(input):
        for iy, char in enumerate(line):
            if char == "^":
                return "up", ix, iy
            elif char == ">":
                return "right", ix, iy
            elif char == "v":
                return "down", ix, iy
            elif char == "<":
                return "left", ix, iy
```

```python
def go_up(direction, ix, iy, visited, input):
    if direction == "up":
        for x_position in range(ix, -1, -1):
            if input[x_position][iy] == "#":
                ix, iy = visited[-1][0], visited[-1][1]
                direction = "right"
                break
            elif input[x_position][iy] != "#":
                visited.append((x_position,iy))
            elif x_position == 0 or x_position == len(input):
                direction = "end"
                break
    return direction, ix, iy, visited
```

```python
def go_right(direction, ix, iy, visited, input):
    if direction == "right":
        for y_position in range(iy, len(input[0]), 1):
            if input[ix][y_position] == "#":
                ix, iy = visited[-1][0], visited[-1][1]
                direction = "down"
                break
            elif input[ix][y_position] != "#":
                visited.append((ix,y_position))
            elif y_position == 0 or y_position == len(input[0]):
                direction = "end"
                break
    return direction, ix, iy, visited
```

```python
def go_right(direction, ix, iy, visited, input):    
    if direction == "right":
        for y_position in range(iy, len(input[0]), 1):
            if input[ix][y_position] == "#":
                ix, iy = visited[-1][0], visited[-1][1]
                direction = "down"
                break
            elif input[ix][y_position] != "#":
                visited.append((ix,y_position))
            elif y_position == 0 or y_position == len(input[0]):
                direction = "end"
                break
    return direction, ix, iy, visited
```

```python
def go_down(direction, ix, iy, visited, input):
    if direction == "down":
        for x_position in range(ix, len(input), 1):
                if input[x_position][iy] == "#":
                    ix, iy = visited[-1][0], visited[-1][1]
                    direction = "left"
                    break
                elif input[x_position][iy] != "#":
                    visited.append((x_position,iy))
                elif x_position == 0 or x_position == len(input):
                    direction = "end"
                    break
    return direction, ix, iy, visited
```

```python
def go_left(direction, ix, iy, visited, input):
    if direction == "left":
        for y_position in range(iy, -1, -1):
            if input[ix][y_position] == "#":
                ix, iy = visited[-1][0], visited[-1][1]
                direction = "up"
                break
            elif input[ix][y_position] != "#":
                visited.append((ix,y_position))
            elif y_position == 0 or y_position == len(input[0]):
                direction = "end"
                break
    return direction, ix, iy, visited
```

```python
input = read_file("example-01.txt")
direction, ix, iy = start_finder(input)
visited = []
```

```python
new_input = input_lists_modifier(input)
new_input
```

```python
modify_input = input
print(modify_input, input)

for ix, line in enumerate(input):
    for iy, char in enumerate(line):
        if char == ".":
            modify_input[ix][iy] = "0"
        print(modify_input)
        modify_input = input
```

```python
while (0 < ix and ix < (len(input)-1) and 0 < iy and iy < (len(input[0])-1) ):

    direction, ix, iy, visited = go_up(direction, ix, iy, visited, input)
    print("GONE UP")

    direction, ix, iy, visited = go_right(direction, ix, iy, visited, input)
    print("GONE RIGHT")

    direction, ix, iy, visited = go_down(direction, ix, iy, visited, input)
    print("GONE DOWN")

    direction, ix, iy, visited = go_left(direction, ix, iy, visited, input)
    print("GONE LEFT")

    ix = visited[-1][0]
    iy = visited[-1][1]

    print(ix, iy)
```
